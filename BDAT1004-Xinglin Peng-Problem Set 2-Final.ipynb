Question 3: Write a class named Marsupial, and write a class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial and show below

```python
class Marsupial(object):
    def __init__(self):
        self.lst = []
    def put_in_pouch(self, item):
        self.lst.append(item)
    def pouch_contents(self):
        return self.lst
```

```python
>>> m = Marsupial()
>>> m.put_in_pouch('doll')
>>> m.put_in_pouch('firetruck')
>>> m.put_in_pouch('kitten')
>>> m.pouch_contents()
```

```python
class Kangaroo(Marsupial):
    def __init__(self,x,y):
        Marsupial.__init__(self)
        self.x = x
        self.y = y
    def jump(self,dx,dy):
        self.x += dx
        self.y += dy
    def __str__(self):
        return 'I am a Kangaroo located at coordinates ({},{})'.format(self.x,self.y)
```

```python
k = Kangaroo(0,0)
print(k)
```

```python
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
```

```python
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

Question 5:Write function collatz() that takes a positive integer x as input and prints the
Collatz sequence starting at x. 

```python
#Write a function
def collatz(n):
    print(n)
    while n != 1:
        if n % 2 == 0:
            n = int(n /2)
        else:
            n = 3 * n + 1
        print(n)
```

Question 7:Implement a class named HeadingParser that can be used to parse an HTML
document, and retrieve and print all the headings in the document.
References:https://docs.python.org/3/library/html.parser.html
https://stackoverflow.com/questions/13334044/get-the-html-under-a-tag-using-htmlparser-python

```python
#Import Library
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    headerFlag = False
    indent = 0
    
 #Triggered when an opening tag is encountered
    def handle_starttag(self, tag, attrs):
        if tag == "h1":
            self.headerFlag = True
 #Set indent to 0
            self.indent= 0
        elif tag == "h2":
            self.headerFlag = True
 #Set indent to 1
            self.indent=1
        else:
            self.headerFlag = False
            self.indent= 0
 #Triggered when data found
    def handle_data(self, data):
        if self.headerFlag:
            space = " "*self.indent
            print("%s%s" % (space,data))
 #Handle end of a tag
    def handle_endtag(self, tag):
        if tag in ["h1","h2"]:
            self.headerFlag = False
```

Question 8: Implement recursive function webdir() that takes as input: a URL (as a string) and
non-negative integers depth and indent. 
Reference:https://stackoverflow.com/questions/1080411/retrieve-links-from-web-page-using-python-and-beautifulsoup

```python
#Import Libraries
from bs4 import BeautifulSoup, SoupStrainer
import httplib2 #Import from Conda

def getLinks(url):
    res = []
    http = httplib2.Http()
    status, response = http.request(url)
    for link in BeautifulSoup(response, parse_only=SoupStrainer('a'), features="html.parser"):
        if link.has_attr('href'):
            if link['href'][:5] == "http:" or link['href'][:5] == "https":
                res.append(link['href'])
    return res

def webdir(url, depth, indent):

    print(indent*" " + url)
    if depth == 0:
        return
    for i in getLinks(url):
        webdir(i, depth-1, indent+1)
```

Question 10:Write list comprehension expressions that use list words and generate the following
lists

```python
#List words
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
lst1 = [i.upper() for i in words] #a
lst2 = [i.lower() for i in words] #b
lst3 = [len(i) for i in words] #c
lst4 = [[i.upper(),i.lower(),len(i)] for i in words] #d
lst5 = [i.lower() for i in words if len(i)>=4] #e
print (lst1)
print (lst2)
print (lst3)
print (lst4)
print (lst5)
```
